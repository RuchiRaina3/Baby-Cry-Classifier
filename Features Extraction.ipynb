### For Naive Bayes Model, features extracted are:


*   **rms energy:** Root Mean Square of energy is equals to sqrt(sum(energy**2))/len(energy))
*   **zcr:** Zero Cross Rate is the rate at which a signal changes from positive to zero to negative or from negative to zero to positive. In simpler words, it's the rate at which the signal passes through the origin.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import librosa
import math
```

```python
FRAME_LENGTH=1024  #Number of samples in one frame
HOP_LENGTH=FRAME_LENGTH  #It is the step size for range function
energy_list=[]  #List to store energy of each audio file
zcr_list=[]     #List to store zcr of each audio file
label_list = [] #List to store label of each audio file
```

```python
def featuresExtraction(path,label):
  files = librosa.util.find_files(path)

  for file in files:
    audio,sr = librosa.load(file)
    
    shape_of_energy = (len(audio)//FRAME_LENGTH) + 1
    energy = np.zeros(shape_of_energy)
    frame = 0
    
    for i in range (0, len(audio), HOP_LENGTH):
      energy[frame]=(sum(audio[i : i + FRAME_LENGTH]**2))  #Energy of a frame = sum of the square of samples of that frame
      frame = frame + 1
    rms_of_energy = math.sqrt((sum(energy**2))/len(energy))
    zcr = sum(librosa.zero_crossings(audio))
    energy_list.append(rms_of_energy)
    zcr_list.append(zcr)
    label_list.append(label)
```

```python
featuresExtraction('/content/drive/MyDrive/Baby Cry Project/Cry Clips/bellypain',1)
featuresExtraction('/content/drive/MyDrive/Baby Cry Project/Cry Clips/burping',2)
featuresExtraction('/content/drive/MyDrive/Baby Cry Project/Cry Clips/discomfort',3)
featuresExtraction('/content/drive/MyDrive/Baby Cry Project/Cry Clips/hungry',4)
featuresExtraction('/content/drive/MyDrive/Baby Cry Project/Cry Clips/tired',5)
```

### DataFrame Creation

```python
list_df = list(zip(energy_list,zcr_list,label_list))
df=pd.DataFrame(list_df,columns=['Energy','ZCR','Label'])

display(df)
```
